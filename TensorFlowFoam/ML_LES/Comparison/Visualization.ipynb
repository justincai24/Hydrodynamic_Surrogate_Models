```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load DSM data
loc_1 = np.loadtxt('../Dyn_Smag/D_SMA_Case/Loc_1.csv',delimiter=',',skiprows=1)[:,:10]
loc_2 = np.loadtxt('../Dyn_Smag/D_SMA_Case/Loc_2.csv',delimiter=',',skiprows=1)[:,:10]
loc_3 = np.loadtxt('../Dyn_Smag/D_SMA_Case/Loc_3.csv',delimiter=',',skiprows=1)[:,:10]
loc_4 = np.loadtxt('../Dyn_Smag/D_SMA_Case/Loc_4.csv',delimiter=',',skiprows=1)[:,:10]
loc_5 = np.loadtxt('../Dyn_Smag/D_SMA_Case/Loc_5.csv',delimiter=',',skiprows=1)[:,:10]

# Add all and average
dsm_data = 0.5*(loc_1+loc_2+loc_3+loc_4+loc_5)
```

```python
#load ML data
loc_1 = np.loadtxt('../ML_LES/ML_LES_Case/Loc_1.csv',delimiter=',',skiprows=1)[:,:10]
loc_2 = np.loadtxt('../ML_LES/ML_LES_Case/Loc_2.csv',delimiter=',',skiprows=1)[:,:10]
loc_3 = np.loadtxt('../ML_LES/ML_LES_Case/Loc_3.csv',delimiter=',',skiprows=1)[:,:10]
loc_4 = np.loadtxt('../ML_LES/ML_LES_Case/Loc_4.csv',delimiter=',',skiprows=1)[:,:10]
loc_5 = np.loadtxt('../ML_LES/ML_LES_Case/Loc_5.csv',delimiter=',',skiprows=1)[:,:10]

# Add all and average
ml_data = 0.5*(loc_1+loc_2+loc_3+loc_4+loc_5)
```

```python
#load UNS data
loc_1 = np.loadtxt('../UNS_Validation/Loc_1.csv',delimiter=',',skiprows=1)[:,:10]
loc_2 = np.loadtxt('../UNS_Validation/Loc_2.csv',delimiter=',',skiprows=1)[:,:10]
loc_3 = np.loadtxt('../UNS_Validation/Loc_3.csv',delimiter=',',skiprows=1)[:,:10]
loc_4 = np.loadtxt('../UNS_Validation/Loc_4.csv',delimiter=',',skiprows=1)[:,:10]
loc_5 = np.loadtxt('../UNS_Validation/Loc_5.csv',delimiter=',',skiprows=1)[:,:10]

# Add all and average
uns_data = 0.5*(loc_1+loc_2+loc_3+loc_4+loc_5)
```

```python
# Make plots - Reynolds stresses
plt.figure()
plt.plot(dsm_data[:,4],label='Dyn. Smagorinsky',linewidth=4)
plt.plot(ml_data[:,4],'ro',markersize=3,label='ML')
plt.plot(uns_data[:,4],'bo',markersize=3,label='No LES model')
plt.legend()
plt.title('Rxx')
plt.show()

plt.figure()
plt.plot(dsm_data[:,5],label='Dyn. Smagorinsky')
plt.plot(ml_data[:,5],'ro',markersize=3,label='ML')
plt.plot(uns_data[:,5],'bo',markersize=3,label='No LES model')
plt.legend()
plt.title('Ryy')
plt.show()

plt.figure()
plt.plot(dsm_data[:,6],label='Dyn. Smagorinsky')
plt.plot(ml_data[:,6],'ro',markersize=3,label='ML')
plt.plot(uns_data[:,6],'bo',markersize=3,label='No LES model')
plt.legend()
plt.title('Rzz')
plt.show()

plt.figure()
plt.plot(dsm_data[:,7],label='Dyn. Smagorinsky')
plt.plot(ml_data[:,7],'ro',markersize=3,label='ML')
plt.plot(uns_data[:,7],'bo',markersize=3,label='No LES model')
plt.legend()
plt.title('Rxy')
plt.show()
```

```python
# Quantitative - Rxx
ml_error = np.sum(np.abs(dsm_data[:,4]-ml_data[:,4]))
uns_error = np.sum(np.abs(dsm_data[:,4]-uns_data[:,4]))
print('Error due to ML:',ml_error, ' Error due to no model:', uns_error)

# Quantitative - Ryy
ml_error = np.sum(np.abs(dsm_data[:,5]-ml_data[:,5]))
uns_error = np.sum(np.abs(dsm_data[:,5]-uns_data[:,5]))
print('Error due to ML:',ml_error, ' Error due to no model:', uns_error)

# Quantitative - Rzz
ml_error = np.sum(np.abs(dsm_data[:,6]-ml_data[:,6]))
uns_error = np.sum(np.abs(dsm_data[:,6]-uns_data[:,6]))
print('Error due to ML:',ml_error, ' Error due to no model:', uns_error)

# Quantitative - Rxy - important
ml_error = np.sum(np.abs(dsm_data[:,7]-ml_data[:,7]))
uns_error = np.sum(np.abs(dsm_data[:,7]-uns_data[:,7]))
print('Error due to ML:',ml_error, ' Error due to no model:', uns_error)
```
