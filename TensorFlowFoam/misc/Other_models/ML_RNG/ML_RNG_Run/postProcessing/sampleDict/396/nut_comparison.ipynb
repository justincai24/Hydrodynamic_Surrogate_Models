```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fname = 'x_by_h_01_nut_ML_nut_RNG.csv'
data = np.loadtxt(fname,delimiter=',',skiprows=1)
print(np.shape(data))
```

```python
plt.figure()
plt.plot(data[:,1],data[:,0],'o',label=r'$\nu_t$-ML',markersize=4,fillstyle='none')
plt.plot(data[:,2],data[:,0],label=r'$\nu_t$-RNG')
plt.xlabel('Magnitude')
plt.ylabel('Height')
plt.legend()
plt.show()
```
