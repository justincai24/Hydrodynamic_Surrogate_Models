```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fname = 'x_by_h_01_U_ML_U_KO.csv'
data = np.loadtxt(fname,delimiter=',',skiprows=1)
print(np.shape(data))
```

```python
plt.figure()
plt.plot(data[:,1],data[:,0],label='ML',marker='o',markersize=4,fillstyle='none')
plt.plot(data[:,4],data[:,0],label='KO')
plt.xlabel('Ux')
plt.ylabel('Height')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.plot(data[:,2],data[:,0],'bo',label='ML',markersize=4)
plt.plot(data[:,5],data[:,0],label='KO')
plt.xlabel('Uy')
plt.ylabel('Height')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.plot(np.sqrt(data[:,1]**2+data[:,2]**2),data[:,0],marker='o',markersize=4,fillstyle='none',label='r$|U|$-ML')
plt.plot(np.sqrt(data[:,4]**2+data[:,5]**2),data[:,0],label='KO')
plt.xlabel('U Magnitude')
plt.ylabel('Height')
plt.legend()
plt.show()
```
